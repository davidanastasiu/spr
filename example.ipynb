```python
# set up algorithm options
from utils.options import Options
options = Options(settings='settings.yml')

# load the data
from utils.data_loader import DataLoader
dl = DataLoader(options)
```

```python
# train an SPR model with Histogram_Boosted_Trees estimators
from spr import SPR
model = SPR('HBT', dl)
model.fit()
```

```python
# predict test samples
status_predictions, concentration_predictions = model.predict(x=dl.X_test)
list(zip(status_predictions, concentration_predictions))
```

```python
# or simply evaluate the model on the test set
df, metrics = model.eval()
metrics
```
